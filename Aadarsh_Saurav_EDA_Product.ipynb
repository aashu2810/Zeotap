```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
  # Replace this with the path to your dataset file
products_data = pd.read_csv('Products.csv')

# Display the first few rows of the dataset to understand its structure
print("First 5 rows of the dataset:")
print(products_data.head())

# Check for missing values in the dataset
print("\nMissing values in each column:")
print(products_data.isnull().sum())

```

```python
print("\nUnique categories and their counts:")
category_counts = products_data['Category'].value_counts()
print(category_counts)

# Plotting category distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
print("\nStatistical summary of product prices:")
print(products_data['Price'].describe())

# Plotting price distribution
plt.figure(figsize=(8, 5))
sns.histplot(products_data['Price'], bins=20, kde=True, color='blue')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
print("\nAverage price per category:")
avg_price_per_category = products_data.groupby('Category')['Price'].mean()
print(avg_price_per_category)

# Plotting average price per category
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_per_category.index, y=avg_price_per_category.values, palette='coolwarm')
plt.title('Average Price per Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
print("\nPrice range (min and max) by category:")
price_range_per_category = products_data.groupby('Category')['Price'].agg(['min', 'max'])
print(price_range_per_category)

# Visualizing price ranges with a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Price', data=products_data, palette='Set3')
plt.title('Price Ranges by Category')
plt.xlabel('Category')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```
